# Task 2
Deal with null values, imputing the mean for numeric features and the string unknown for categorical features

```python
from sklearn.impute import SimpleImputer

df3 = df2.copy()

categ_feat = ["F13", "class"]
numeric_feat = list(df3.columns)
for c in categ_feat:
    numeric_feat.remove(c)

print(f"the numeric features are: {numeric_feat}")
print(f"the categorical features are: {categ_feat}")

imputer_num = SimpleImputer(strategy="mean")
imputer_categ = SimpleImputer(strategy="constant", fill_value="unknown")

df3[numeric_feat] = imputer_num.fit_transform(df3[numeric_feat])
df3[categ_feat] = imputer_categ.fit_transform(df3[categ_feat])
df3.sample(5)
```

```python
df3.describe(include="all")
```

# Task 3
train, optimize and test two classifier models of your choice, the optimization must be done with cross validation, optimize the f1-score_macro

```python
# droppo la feature F13 in quanto categorica e non compatibile con i classificatori che voglio usare
df4 = df3.drop("F13", axis=1)
df4.sample(5)
```

```python
# dividiamo il dataset in training set e test set. Faremo crossvalidation sul training set
from sklearn.model_selection import train_test_split

target = "class"
X = df4.drop(target, axis=1)
y = df4[target].astype("int") # importante per riuscire a fittare i modelli

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
print(f"Dimesioni training set: X_train={X_train.shape}; y_train={y_train.shape}")
print(f"Dimesioni test set: X_test={X_test.shape}; y_test={y_test.shape}")
```

```python
corr = X.corr()
plt.figure(figsize=(15,15))
sns.heatmap(corr, annot=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

scoring = "f1_macro"

model_labels = ["dt", "knn"]

models = {
    "dt": {
        "name": "Decision Tree",
        "estimator": DecisionTreeClassifier(),
        "parameters": {
            "max_depth": list(range(2,11))
        }
    },
    "knn": {
        "name": "K-Nearest Neighbour",
        "estimator": KNeighborsClassifier(),
        "parameters": {
            "n_neighbors": list(range(2,5))
        }
    }
}
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

results = pd.DataFrame(columns=["estimator", "best-params", "f1-score"])
tuned_models = []

n_folds = 3
for ml in model_labels:
    grid = GridSearchCV(
        models[ml]["estimator"],
        models[ml]["parameters"],
        scoring=scoring,
        cv=n_folds
    )

    grid.fit(X_train, y_train)
    tuned_models.append(grid.best_estimator_)
    y_pred = grid.best_estimator_.predict(X_test)
    report = classification_report(y_test, y_pred, output_dict=True)
    
    print(f"report for {models[ml]["name"]}")
    print(classification_report(y_test, y_pred, zero_division=0))

    results.loc[len(results)] = [models[ml]["name"], grid.best_params_, report["macro avg"]["f1-score"]]
```

```python
from sklearn.feature_selection import SelectKBest, mutual_info_classif

print(f"prima di fare feature selection: X_train, y_train, X_test, y_test = {X_train.shape}, {y_train.shape}, {X_test.shape}, {y_test.shape}")

kb = SelectKBest(mutual_info_classif, k=10)
X_train_transformed = kb.fit_transform(X_train, y_train) # decido quali feature sono importanti guardando il train set
X_test_transformed = kb.transform(X_test)                # uso le stesse feature anche per il test set 

print(f"dopo aver fatto feature selection: X_train, y_train, X_test, y_test = {X_train_transformed.shape}, {y_train.shape}, {X_test_transformed.shape}, {y_test.shape}")
```

```python
results_fs = pd.DataFrame(columns=["estimator", "best-params", "f1-score"])
tuned_models_fs = []

n_folds = 3
for ml in model_labels:
    grid = GridSearchCV(
        models[ml]["estimator"],
        models[ml]["parameters"],
        scoring=scoring,
        cv=n_folds
    )

    grid.fit(X_train_transformed, y_train)
    tuned_models_fs.append(grid.best_estimator_)
    y_pred = grid.best_estimator_.predict(X_test_transformed)
    report = classification_report(y_test, y_pred, output_dict=True)
    
    print(f"report for {models[ml]["name"]}")
    print(classification_report(y_test, y_pred, zero_division=0))

    results_fs.loc[len(results_fs)] = [models[ml]["name"], grid.best_params_, report["macro avg"]["f1-score"]]
```
