# Task 2
Eliminate:
- totally null columns
- totally null rows 
- columns with less than 1/3 of non null values

fill the remaining NaN values with the mean of the column

```python
from sklearn.impute import SimpleImputer

df2 = df.copy()
df2 = df2.dropna(axis=0, how="all") # dropping all-null rows
df2 = df2.dropna(axis=1, how="all") # dropping all-null columns
thresh = df2.shape[0] // 3 # minimum number of non-null column values
df2 = df2.dropna(axis=1, thresh=thresh) # dropping columns with less then 1/3 of non null values
```

```python
df2.info()
```

```python
df2.select_dtypes(include="number").columns
```

```python
imputer = SimpleImputer(strategy='mean')
# non applico la trasformazione alle prime due colonne in quanto non hanno missing values e in quanto non funziona su stringhe
df2[df2.columns[2:]] = imputer.fit_transform(df2[df2.columns[2:]]) 
df2.describe(include="all")

# sarebbe meglio fare cos√¨
# col2transform = df2.select_dtypes(include="number").columns
# df2[col2transform] = imputer.fit_transform(df2[df2[col2transform]]) 
```

# Task 3
Drop Time, convert Date from string to datetime and group by Date using mean as aggregate function

```python
df3 = df2.copy()
df3 = df3.drop("Time", axis=1)
df3["Date"] = pd.to_datetime(df3["Date"], dayfirst=True)
df3 = df3.groupby("Date").mean()
df3
```

# Task 4

- Discretise continuous values with two bins, kmeans strategy and onehot-dense encoding
- Discretization/encoding generates 0/1 values; convert the binary values obtained into boolean, as requested by Apriori

```python
from sklearn.preprocessing import KBinsDiscretizer

df4 = df3.copy()

kbd = KBinsDiscretizer(n_bins=2, strategy="kmeans", encode="onehot-dense")
df4 = pd.DataFrame(kbd.fit_transform(df4))
df4
```

```python
def binarizer_function(x):
    if x >= 1:
        return True
    else:
        return False

df4 = df4.map(binarizer_function)
df4
```

# Task 5
Set the names of two columns generated by the discretisation of each attribute A to A_low, A_high 
- (with discretisation/one-hot-encoding, each original column generates two columns, the first is for the low values, the second for the high values)

```python
new_columns = []

for name in df3.columns:
    new_columns.append(name+"_low")
    new_columns.append(name+"_high")

df4.columns = new_columns
df4
```

# Task 6

Find a value of min_support such that the Apriori algorithm generates at least 8 frequent itemsets with at least 2 items, output the result 

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import numpy as np

for min_sup in np.arange(1, 0.00, -0.01):
    itemsets=apriori(df4, min_support=min_sup, use_colnames=True)

    # filtro le righe che contengono un itemset con un solo elemento 
    for i, row in itemsets.iterrows():
        if(len(row["itemsets"]) < 2):
            itemsets = itemsets.drop(i, axis=0)
        
    if len(itemsets) >= 8:
        break

print(f"{min_sup:.2f}")
itemsets
```

# Task 7
Find the minimum metric threshold (quale metric???) such that at least 100 association rules are extracted from the frequent itemsets found and show the metrics (plurale???) used and the best 10 rules by descending confidence and support

```python
for min_thresh in np.arange(1, 0.00, -0.01):
    rules = association_rules(itemsets, metric="confidence", min_threshold=min_thresh)
    if len(rules) >= 100:
        break

print(min_thresh)
rules.sort_values(by=["confidence", "support"], ascending=False)[["antecedents", "consequents", "support", "confidence", "lift"]].head(10)
```
